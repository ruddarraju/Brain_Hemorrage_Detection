```python
import os
import shutil
import random
```

```python
BASE_DIR = r"C:\Users\abhil\Desktop\model_z\Data"
CLASSES = ["Hemorrhagic", "NORMAL"]

TRAIN_DIR = os.path.join(BASE_DIR, "Train")
TEST_DIR  = os.path.join(BASE_DIR, "Test")

SPLIT_RATIO = 0.8
SEED = 42
random.seed(SEED)

IMAGE_EXTENSIONS = (".jpg", ".jpeg", ".png", ".bmp")
```

```python
for split_dir in [TRAIN_DIR, TEST_DIR]:
    for cls in CLASSES:
        os.makedirs(os.path.join(split_dir, cls), exist_ok=True)
```

```python
def collect_images(root_dir):
    image_paths = []
    for root, _, files in os.walk(root_dir):
        for file in files:
            if file.lower().endswith(IMAGE_EXTENSIONS):
                image_paths.append(os.path.join(root, file))
    return image_paths
```

```python
for cls in CLASSES:
    class_dir = os.path.join(BASE_DIR, cls)
    images = collect_images(class_dir)

    random.shuffle(images)
    split_point = int(len(images) * SPLIT_RATIO)

    train_images = images[:split_point]
    test_images  = images[split_point:]

    for img_path in train_images:
        shutil.copy(
            img_path,
            os.path.join(TRAIN_DIR, cls, os.path.basename(img_path))
        )

    for img_path in test_images:
        shutil.copy(
            img_path,
            os.path.join(TEST_DIR, cls, os.path.basename(img_path))
        )

    print(f"{cls}: {len(train_images)} train, {len(test_images)} test")
```

```python
print("✅ Nested dataset split completed successfully")
```
